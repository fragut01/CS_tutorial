```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
import matplotlib.image as mpimg
import os
```

```python
LOLA = mpimg.imread(os.path.join('C:\\Users\\SE79065\\Documents\\CompressiveSensing\\images\\LOLA.jpg'))
LOLA = np.mean(LOLA,-1)
```

```python
lola_trans = np.abs(np.fft.fft2(LOLA))
x,y = np.shape(LOLA)
```

```python
mag_spect = 20*np.log(lola_trans)
```

```python
sup = mag_spect.max()
inf = mag_spect.min()
plt.imshow(mag_spect,cmap='seismic',vmax = sup,vmin=inf,aspect='auto')
plt.colorbar()
```
